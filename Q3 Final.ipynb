```python
import math 
import string 
import sys 
  
# reading the text file 
# This functio will return a  
# list of the lines of text  
# in the file. 
def read_file(filename):  
      
    try: 
        with open(filename, 'r') as f: 
            data = f.read() 
        return data 
      
    except IOError: 
        print("Error opening or reading input file: ", filename) 
        sys.exit() 
  
# splitting the text lines into words 
# translation table is a global variable 
# mapping upper case to lower case and 
# punctuation to spaces 
#translation_table = str.maketrans(string.punctuation+string.ascii_uppercase, " "*len(string.punctuation)+string.ascii_lowercase)

# returns a list of the words 
# in the file 
def make_list(document):  
   
    #text = text.translate(translation_table) 
    mylist = document.split()
    
    newlist=[]
    
    for text in mylist:
        text=text.lower()
        for punc in text:
            if punc in [",",".","'","!","?",":",";","-","(",")"]:
                text=text.replace(punc,"")
        newlist.append(text)

    return newlist
  
  
# counts frequency of each word 
# returns a dictionary which maps 
# the words to  their frequency. 
def count_freq(word_list):  
      
    D = {} 
      
    for new_word in word_list: 
          
        if new_word in D: 
            D[new_word] = D[new_word] + 1
              
        else: 
            D[new_word] = 1
              
    return D 
  
  
  
# returns the dot product of two documents 
def dot_product(D1, D2):  
    Sum = 0.0
      
    for key in D1: 
          
        if key in D2: 
            Sum += (D1[key] * D2[key]) 
              
    return Sum

def MultiDoc(D1, Dlist=[]):
    
    D1=read_file(D1)
    D1_list=make_list(D1)
    D1_freq=count_freq(D1)
    file_list=[]        
    products=[]
    for doc in Dlist:
        D2=read_file(doc)
        D2_list=make_list(D2)
        D2_freq=count_freq(D2_list)
        file_list.append(doc)
        products.append(dot_product(D1_freq,D2_freq))
    result=list(zip(products, file_list))
    result.sort(reverse=True)
    result_sorted=[x for y, x in result]   #[file_list for products, file_list in result]
    
    return result_sorted

      

Doc="/Users/astrohonova/Desktop/TestDoc.txt"
DocTwo="/Users/astrohonova/Desktop/DocTwo.txt"
DocThree="/Users/astrohonova/Desktop/DocThree.txt"

TestDoc=read_file(Doc)
TestDocTwo=read_file(DocTwo)
TestDocThree=read_file(DocThree)

wordlist1=make_list(TestDoc)
wordlist2=make_list(TestDocTwo)
wordlist3=make_list(TestDocThree)

freq1=count_freq(wordlist1)
freq2=count_freq(wordlist2)
freq3=count_freq(wordlist3)
print(freq1)


print(dot_product(freq1, freq2))
print(dot_product(freq1, freq3))

print(MultiDoc(Doc, [DocTwo, DocThree]))
```
